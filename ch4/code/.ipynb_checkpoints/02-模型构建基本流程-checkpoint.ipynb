```python
# 导入相应模块
from sklearn.linear_model import LogisticRegression # 逻辑回归算法
from sklearn.model_selection import train_test_split # 用于交叉验证数据拆分
from sklearn.metrics import classification_report #模型评估
```

```python
# 导入数据预处理函数
%run 03-数据预处理函数.ipynb
```

```python
# step-1：数据预处理
TrainData = prepare_data("datasets/titanic/train.csv","train")
TrainData.head()
```

```python
TrainData.describe().transpose()
```

```python
# step2：模型训练

# 拆分特征根与目标变量
trainData_X = TrainData.drop(['Survived'], axis = 1)
trainData_y = TrainData.Survived 

# 拆分训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(trainData_X, trainData_y, test_size=0.3, random_state=123456)

# 模型实例化
lr = LogisticRegression()

# 训练模型
lr.fit(X_train, y_train)

# step3：模型评估
# 对剩余的30%数据样本做预测
y_test_pred = lr.predict(X_test)
# 比较预测值与真实值
print(classification_report(y_test,y_test_pred))
```

```python
# 处理新数据样本
TestData = prepare_data("datasets/titanic/test.csv","test")
# 准备测试集，不需要ID
Test_X = TestData.drop(['PassengerId'], axis = 1)
# 对测试数据进行预测
pred = lr.predict(Test_X)
# 生成最终提交结果Submission 
Submission = pd.DataFrame({'PassengerId':TestData.PassengerId,'Survived':pred})
# 导出为csv文件
Submission.to_csv("./datasets/titanic/Submission.csv", index=False)
# 查看导出文件，读取测试结果数据集
submission_df = pd.read_csv('./datasets/titanic/Submission.csv')
submission_df.head()
```
